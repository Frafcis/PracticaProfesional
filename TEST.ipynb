```python
import requests
import json
import pandas as pd
```

```python
def get_info(data):
    info = []
    for x in range(0,len(data)-1):
        info.append(data["info.authors.author"].iloc[x])
    return info
```

```python
response = requests.get("https://dblp.org/search/publ/api?q=test&h=1000&format=json")
print(response.json())
```

```python
dictr = response.json()
hits =  dictr['result']['hits']['hit']

# info = get_info(hits)

hits_df = pd.json_normalize(hits)
hits_df = hits_df[hits_df["info.year"] >= "2000"]
hits_df = hits_df[hits_df["info.authors.author"].notna()]
hits_df = hits_df[hits_df["info.type"] == "Journal Articles"]

info = get_info(hits_df)
```
